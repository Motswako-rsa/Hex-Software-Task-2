```python
sns.pairplot(data, hue='Species', diag_kind='kde')
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

```python
data.drop(columns='Id', inplace=True, errors='ignore')
data.hist(edgecolor='black', figsize=(10,6))
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
data.drop(columns='Id', inplace=True, errors="ignore")
sns.boxplot(data=data, orient="h", palette="Set2")
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(hue="Species", x="Species", y="PetalLengthCm", data=data, legend=False, palette="muted")
plt.title("Violin Plot of Petal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = data.drop('Species', axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species", data=data, palette="bright")
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```
